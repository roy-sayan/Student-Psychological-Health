```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the CSV data
df = pd.read_csv('Data Carrard et al. 2022 MedTeach.csv')
df
```

```python
# Set up the data variables
gender = df['sex']
job_satisfaction = df['jspe']
cognitive_empathy = df['qcae_cog']
affective_empathy = df['qcae_aff']
```

```python
# Calculate the correlations and store them in a list
corr_matrix = [gender.corr(job_satisfaction), cognitive_empathy.corr(job_satisfaction), affective_empathy.corr(job_satisfaction)]
```

```python
# select the columns to analyze
cols = ["sex", "jspe", "qcae_cog", "qcae_aff"]
```

```python
# Plot the correlations on a line graph
plt.plot(corr_matrix)
plt.xlabel('Gender, Cognitive Empathy, Affective Empathy')
plt.ylabel('Job Satisfaction Correlation')
plt.title('Correlation between Job Satisfaction and Gender, Cognitive Empathy, Affective Empathy')
plt.show()
```

```python
# obtain the correlation matrix
corr = df.corr()

# create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8,8))
sns.heatmap(corr,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap="Oranges")
```
